```python
def update(x):
    if x == 'M':
        return 1
    else:
        return 0


df['target'] = df['R'].apply(update)
```

```python
df['target']
```

```python
df.drop('R',axis=1,inplace=True)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target',axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix

lg = LogisticRegression()
lg.fit(X_train,y_train)
lg_predict = lg.predict(X_test)

print(confusion_matrix(y_test,lg_predict))
print(classification_report(y_test,lg_predict))
```

# Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_predict = rf.predict(X_test)

print(confusion_matrix(y_test,rf_predict))
print(classification_report(y_test,rf_predict))
```

# Cross Validation Score 

```python
from sklearn.model_selection import cross_val_score

score = cross_val_score(RandomForestClassifier(),X,y,cv=10,scoring='accuracy')

print(f'Mean Score -> {score.mean()}')
print()
print(f'Max Score -> {score.max()}')
```

# Using PCA ....

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)
```
